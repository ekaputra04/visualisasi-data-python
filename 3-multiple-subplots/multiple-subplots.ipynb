Multiple Subplots

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Simple line plot

```python
x = np.arange(0., 2., 0.01)
x
```

```python
s = np.sin(2 * np.pi *x)
s
```

OO Style

```python
fig, ax = plt.subplots() # OO Style
ax.plot(x, s)
ax.set(
  xlabel='nilai x',
  ylabel='nilai sin(x)',
  title='Visualisasi nilai sin'
)
ax.grid()
plt.show()
```

pyplot style

```python
plt.plot(x, s)
plt.xlabel('nilai x')
plt.ylabel('nilai sin(x)')
plt.title('Visualisasi nilai sin')
plt.legend()
plt.grid()
```

Multiple Subplots

```python
x1 = np.linspace(0., 5., 100)
x2 = np.linspace(0., 2., 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x2
```

```python
y1
```

```python
y2
```

OO Style

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai $cos(x_1)$')

ax2.plot(x2, y2, 'g-')
ax2.set_ylabel('Nilai $cos(x_2)$')

ax2.set_xlabel('Nilai x')

ax1.grid()
ax2.grid()
plt.show()
```

pyplot Style

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')
plt.grid()

plt.show()
```
