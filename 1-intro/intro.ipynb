Membuat plotting sederhana

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

fig, ax = plt.subplots()
ax.plot(x, y)
```

Alternatif lain

```python
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

plt.plot(x, y)
```

Ada 2 cara menggunakan matplotlib:
- OO Style (Object Oriented Style)
- pyplot Style

Mengenal OO Style untuk melakukan plotting

```python
import numpy as np

x = np.linspace(0, 2, 100)
x
```

```python
# membuat figure dan axes
fig, ax = plt.subplots() 

# plotting 3 varian data pada axes
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='kubic')

ax.set_xlabel('sumbu x') # menyertakan label pada x axis
ax.set_ylabel('sumbu y') # menyertakan label pada y axis
ax.set_title('Simple Plot') # menyertakan title pada axes
ax.legend() # menyertakan legend
```

Menggunakan pyplot Style untuk melakukan plotting

```python
# plotting 3 varian data pada axes
# plotting 3 varian data pada axes
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='kubic')

plt.xlabel('sumbu x') # menyertakan label pada x axis
plt.ylabel('sumbu y') # menyertakan label pada y axis
plt.title('Simple Plot') # menyertakan title pada axes
plt.legend() # menyertakan legend
```
