3D Area Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
```

Sample Dataset

```python
n = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

3D Wireframe

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='orange')

ax.set_xlabel('Sumbu x')
ax.set_ylabel('Sumbu y')
ax.set_zlabel('Sumbu z')
ax.set_title('3D Wireframe')

plt.show()
```

3D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# ax.plot_surface(x, y, z, color='orange')
ax.plot_surface(x, y, z, color='orange', cmap='viridis')

ax.set_xlabel('Sumbu x')
ax.set_ylabel('Sumbu y')
ax.set_zlabel('Sumbu z')
ax.set_title('3D Wireframe')

plt.show()
```
