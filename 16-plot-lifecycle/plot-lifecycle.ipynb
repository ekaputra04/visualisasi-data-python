Plot Lifecycle

```python
import matplotlib.pyplot as plt
import numpy as np
```

Sample Dataset

```python
data = {
  'Item A':109283,
  'Item B':192749,
  'Item C':154839,
  'Item D':143756,
  'Item E':127463,
  'Item F':103584,
  'Item G':129483,
  'Item H':146372,
  'Item I':175637,
  'Item J':142637
}
data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

Simple Plot

```python
fig, ax = plt.subplots()

ax.barh(items, count)
ax.grid()

plt.show()
```

Pengaturan Style

```python
plt.style.available
```

```python
plt.style.use('classic')

fig, ax = plt.subplots()

ax.barh(items, count)
ax.grid()

plt.show()
```

Pengaturan Label pada Sumbu (axis) dan Judul (title)

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(
  title='Contoh Plot',
  xlabel='Jumlah',
  ylabel='Items'
)

plt.show()
```

Pengaturan Tick Label

```python
fig, ax = plt.subplots()

ax.barh(items, count)
ax.grid()

labels = ax.get_xticklabels()

# set properties
plt.setp(
  labels,
  rotation=45, # berlawanan arah jarum jam
  horizontalalignment='right'
)

plt.show()
```

Pengaturan Format pada Ticker

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x, pos):
  return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(
  title='Contoh Plot',
  xlabel='Jumlah',
  ylabel='Items'
)

plt.show()
```

Penambahan Garis (vertikal / horizontal) pada Plot

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(
  title='Contoh Plot',
  xlabel='Jumlah',
  ylabel='Items'
)

ax.axvline(
  np.mean(count), # pada koordinat berapa garis terbentuk
  ls='--', # line style
  color='red' # warna
)
plt.show()
```

Menyimpan Hasil Plot ke dalam Suatu File

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(
  title='Contoh Plot',
  xlabel='Jumlah',
  ylabel='Items'
)

# file akan disimpan pada directory yang sama dengan file .ipynb
fig.savefig(
  'sales.png', # nama file
  transparent=True, # latar belakang transparan
  dpi=80 # kepadatan pixel
)
```
