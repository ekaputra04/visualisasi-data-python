Visualisasi Data yang Tersimpan pada Pandas Data Frame

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Sample Dataset

```python
df = pd.DataFrame({
  'nama': ['Cecep', 'Karti', 'Bejo', 'Tejo', 'Asep', 'Wati', 'Amat'],
  'usia': [23, 78, 22, 19, 45, 33, 20],
  'gender': ['L', 'P', 'L', 'L', 'L', 'P', 'L'],
  'provinsi': ['Jawa Timur', 'Jawa Barat', 'Jawa Barat', 'Jawa Timur', 'Jawa Tengah', 'Bali', 'Bali'],
  'n_anak': [3, 2, 1, 1, 3, 2, 2],
  'n_pets': [0, 2, 1, 4, 2, 1, 2]
})

df
```

Scatter Plot

```python
df.plot(kind='scatter', x='n_anak', y='n_pets', color='red')

plt.title('Scatter Plot')
plt.xlabel('Jumlah Anak')
plt.ylabel('Jumlah Pets')
plt.show()
```

Histogram

```python
df[['usia']].plot(kind='hist', bins=[0, 20, 40, 60, 80, 100])

plt.title('Histogram')
plt.xlabel('Usia')
plt.ylabel('Jumlah')
plt.show()
```

Line Plot

```python
df.plot(kind='line', x='nama', y='n_anak', color='green')

plt.title('Line Plot')
plt.xlabel('Nama')
plt.ylabel('Jumlah Anak')
plt.show()
```

Multiple Line Plot

```python
ax = plt.gca()

df.plot(kind='line', x='nama', y='n_anak', color='green', ax=ax)
df.plot(kind='line', x='nama', y='n_pets', color='red', ax=ax)

plt.title('Multiple Line Plot')
plt.xlabel('Nama')
plt.ylabel('Jumlah')
plt.show()
```

Bar Plot

```python
df.plot(kind='bar', x='nama', y='usia')

plt.title('Bar Plot')
plt.xlabel('Nama')
plt.ylabel('Usia')
plt.show()
```

Bar Plot berdasarkan kelompok tertentu

```python
df.groupby('provinsi')['nama'].count().plot(kind='bar')

plt.title('Multiple Line Plot')
plt.xlabel('Nama')
plt.ylabel('Jumlah')
plt.show()
```

Stacked Bar Plot dengan Dua Level Group By

```python
df.groupby(['provinsi', 'gender']).size().unstack().plot(kind='bar', stacked=True)

plt.title('Stacked Bar Plot')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah')
plt.show()
```

```python
df.groupby(['gender', 'provinsi']).size().unstack().plot(kind='bar', stacked=True)

plt.title('Stacked Bar Plot')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah')
plt.show()
```

Bar Plot

```python
data = {
  'negara': ['United States', 'Singapore', 'Germany', 'United Kingdom', 'Japan'] ,
  'gdp': [62686, 100345, 52559, 45705, 44227]
}

df = pd.DataFrame(data, columns=['negara', 'gdp'])
df
```

```python
df.plot(x='negara', y='gdp', kind='bar')

plt.title('Bar Plot')
plt.xlabel('Negara')
plt.ylabel('GDP')
plt.show()
```

Horizontal Bar Plot

```python
df.plot(x='negara', y='gdp', kind='barh')

plt.title('Bar Horizontal Plot')
plt.xlabel('GDP')
plt.ylabel('Negara')
plt.show()
```

Scatter Plot

```python
data = {
  'negara': ['United States', 'Singapore', 'Germany', 'United Kingdom', 'Japan'] ,
  'gdp': [62686, 100345, 52559, 45705, 44227],
  'life_expectancy': [79.24, 82.84, 80.64, 81.40, 83.62]
}

df = pd.DataFrame(data, columns=['negara', 'gdp', 'life_expectancy'])
df
```

```python
df.plot(x='gdp', y='life_expectancy', kind='scatter')

plt.title('Scatter Plot')
plt.xlabel('GDP dan USD')
plt.ylabel('Life Expectancy')
plt.show()
```

Pie Plot

```python
data = {
  'negara': ['United States', 'Singapore', 'Germany', 'United Kingdom', 'Japan'] ,
  'gdp': [62686, 100345, 52559, 45705, 44227]
}

df = pd.DataFrame(data, columns=['negara', 'gdp'])
df = df.set_index('negara') # untuk mengubah agar index bukan angka melainkan kolom tertentu
df
```

```python
df.plot(kind='pie', y='gdp', figsize=(6, 6))
plt.title('GDP')
plt.show()
```

Box Plot
