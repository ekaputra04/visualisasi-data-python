Matplotlib Style

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

Fungsi Plotting

```python
def gambar_barplot(style='Grouped Bar Plot'):
    data1 = [25, 85, 75, 40, 60]
    data2 = [40, 35, 20, 55, 10]
    kategori = ['A', 'B', 'C', 'D', 'E']

    x = np.arange(len(kategori))

    width = 0.35

    plt.bar(x-width/2, data1, width, label='Data 1')
    plt.bar(x+width/2, data2, width, label='Data 2')

    plt.xticks(x, kategori)

    plt.grid()
    plt.xlabel('Kategori')
    plt.ylabel('Jumlah')
    plt.title(style)

    plt.legend()
    plt.show()
```

```python
gambar_barplot()
```

List Daftar Style

```python
plt.style.available
```

Mengubah Style Matplotlib

```python
banyak_style = len(plt.style.available)
print(f'Banyak Style : {banyak_style}')

for i in range(banyak_style):
    style.use(plt.style.available[i])
    print(f'Style : {plt.style.available[i]}')
    gambar_barplot(plt.style.available[i])
```
