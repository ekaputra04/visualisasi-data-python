```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Membuat plotting sederhana
- jika parameter hanya menyertakan 1 list, maka nilai dari list akan dijadikan sumbu y
- jika parameter terdapat 2 deret bilangan, maka list pertama akan menjadi sumbu x dan list kedua akan menjadi sumbu y

```python
x = [2, 3, 4, 7, 11]
y = [5, 8, 3, 5, 4]

plt.plot(x)
plt.ylabel("sumbu y")
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

Format plot

```python
# parameter ke 4 baerarti bentuk diagram
# defaultnya garis biru 'b-'
# opsi bulat menjadi '..o'
# jika merah bulat menjadi 'ro'
plt.plot(x, y, 'go') 
# axis memerlukan parameter list [xmin, xmax, ymin, ymax]
plt.axis([0, 10, 0, 10]) 
plt.show()
```

Multiplot dalam satu pemanggilan fungsi plot

```python
t = np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'g-')
plt.plot(t, t**3, 'b^')
plt.show()
```

```python
plt.plot(t, t, 'r--',
         t, t**2, 'g-',
         t, t**3, 'b^')
plt.show()
```

Plotting dengan keywords

```python
data = {
  'a': np.arange(50),
  'c': np.random.randint(0, 50, 50),
  'd': np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

Scatterplot

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

Plotting untuk tipe data categorical

```python
names = ['grup a', 'grup b', 'grup c']
values = [1, 10, 100]
```

```python
plt.plot(names, values)
plt.show()
```

```python
plt.scatter(names, values)
plt.show()
```

```python
plt.bar(names, values)
plt.show()
```
