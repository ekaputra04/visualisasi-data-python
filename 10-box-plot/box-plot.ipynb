Box Plot / Box Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(
  loc=100, # mengatur nilai tengah
  scale=10, # mengatur standar deviasi
  size=200 # mengatur jumlah data point
)
data
```

Simple Box Plot

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.grid()
plt.show()
```

Notched Box Plot

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.grid()
plt.show()
```

Pengaturan Marker untuk Outlier

```python
plt.boxplot(
  data, 
  labels=['Data'],
  showfliers=True,
  flierprops={
    'markerfacecolor':'r',
    'marker':'d' # d artinya diamond
  }
)

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.grid()
plt.show()
```

Horizontal Box Plot

```python
plt.boxplot(data, labels=['Data'], vert=False)

plt.title('Horizontal Box Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.grid()
plt.show()
```

Multiple Box Plot

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4']

# fungsi reverse untuk membalik posisi 

data.reverse()
labels.reverse()
```

```python
plt.boxplot(data, labels=labels, vert=False) 
plt.title('Multiple Box Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.grid()
plt.show()
```
