Pie Plot / Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(
  data,
  labels=kategori, # opsional
  autopct='%1.1f%%', # untuk menampilkan persentase (opsional)
  startangle=90 # melakukan rotasi (opsional)
)
plt.title('Contoh Pie Chart')
plt.show()
```

Pengaturan Legend dan Color

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(
  data,
  labels=kategori, # opsional
  autopct='%1.1f%%', # untuk menampilkan persentase (opsional)
  startangle=90, # melakukan rotasi (opsional)
  colors=warna
)
plt.legend()
plt.title('Contoh Pie Chart')
plt.show()
```

Exploded Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']

# nilai 0 jika ingin menyatukan chart
# jika nilai > 0 maka bagian chart akan dijauhkan dari pusat
# semakin besar nilai maka semakin menjauh
explode_var = [0., 0., 0.2, 0., 0.08]
```

```python
plt.pie(
  data,
  labels=kategori, # opsional
  autopct='%1.1f%%', # untuk menampilkan persentase (opsional)
  startangle=90, # melakukan rotasi (opsional)
  colors=warna,
  explode=explode_var
)
plt.legend()
plt.title('Contoh Exploded Pie Chart')
plt.show()
```
