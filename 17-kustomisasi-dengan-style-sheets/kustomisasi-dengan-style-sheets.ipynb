Kustomisasi Matplotlib dengan Style Sheets

```python
import matplotlib.pyplot as plt
import numpy as np
```

Predefined Styles

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.title('Penerapan Style')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

plt.show()
```

Membuat Style Sendiri

```python
plt.style.use('./style.mplstyle')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.title('Penerapan Style Sendiri')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

plt.show()
```

Menerapkan Multiple Style

```python
plt.style.use(['classic', 'dark_background'])
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.title('Penerapan Multiple Style')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

plt.show()
```

Menerapkan Temporary Styling

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'ro-')

plt.show()
```

```python
with plt.style.context('dark_background'):
  plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'ro-')

plt.show()
```
