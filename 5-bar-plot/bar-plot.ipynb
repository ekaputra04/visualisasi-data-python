Bar Plot / Bar Chart

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Simple Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

Pengaturan Grid dan Color

```python
plt.bar(
  kategori, 
  data,
  color='red',
  alpha=0.75
)

plt.grid(
  linestyle='--',
  linewidth=2,
  axis='y', # pilihan: x, y, both
  alpha=0.5
)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

Horizontal Bar Plot

```python
# plt.barh(kategori, data)

# untuk membalik urutan:
plt.barh(kategori[::-1], data[::-1])


plt.grid(
  linestyle='--',
  linewidth=2,
  axis='both', # pilihan: x, y, both
  alpha=0.5
)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Bar Plot')

plt.show()
```

Stacked Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data 1')
plt.bar(kategori, data2, label='Data 2', bottom=data1)

# jika tidak dituliskan parameter bottom, maka plotting kedua akan menimpa plotting pertama

plt.legend()
plt.grid()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Stacked Bar Plot')

plt.show()
```

Grouped Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori))
x
```

```python
# semakin kecil width, semakin kecil bar
# maksimal width = 0.5
width = 0.35

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

plt.grid()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Grouped Bar Plot')

plt.legend()
plt.show()
```
