Scatter Plot / Scatter Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

Simple Scatter Plot

```python
x = [2, 4, 6, 8, 10, 11, 11.5, 11.7]
y = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5]
```

```python
plt.scatter(x, y, label='Data 1', color='r')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.title('Contoh Scatter Plot')
plt.legend()
plt.grid()
plt.show()
```

Multiple Scatter Plot

```python
x1 = [2, 4, 6, 8, 10, 11, 11.5, 11.7]
y1 = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5]

x2 = [8, 8.5, 9, 9.5, 10, 10.5, 11]
y2 = [3, 3.5, 3.7, 4, 4.5, 5, 5.2]
```

```python
plt.scatter(x1, y1, label='Data 1', color='r')
plt.scatter(x2, y2, label='Data 2', color='g')

plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.title('Multiple Scatter Plot')
plt.legend()
plt.grid()
plt.show()
```

Pengaturan Marker

```python
x1 = [2, 4, 6, 8, 10, 11, 11.5, 11.7]
y1 = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5]

x2 = [8, 8.5, 9, 9.5, 10, 10.5, 11]
y2 = [3, 3.5, 3.7, 4, 4.5, 5, 5.2]

plt.scatter(
  x1,
  y1,
  color='cyan',
  linewidths=1,
  marker='s', # square
  edgecolors='black',
  s=100 # ukuran marker
)

plt.scatter(
  x2,
  y2,
  color='yellow',
  linewidths=1,
  marker='^', # segitiga
  edgecolors='black',
  s=100 # ukuran marker
)

plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.title('Multiple Scatter Plot')
plt.legend()
plt.grid()
plt.show()
```

Scatter Plot untuk Data Multi Dimensi 

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys()
```

```python
iris['data']
```

```python
iris['feature_names']
```

```python
features = iris['data'].T # .T artinya transpose (membalik baris dan kolom)
features
```

```python
iris['target']
```

```python
iris['target_names']
```

```python
plt.scatter(
  features[0],
  features[1],
  alpha=0.2,
  s=100 * features[3],
  c=iris['target'],
  cmap='viridis'
)

plt.title('Iris Datasets')
plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
plt.colorbar()

plt.show()
```
