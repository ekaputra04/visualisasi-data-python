Violin Plot / Violin Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(
  loc=100, # mean
  scale=10, # standard deviation
  size=200 # banyak data
)
data
```

Simple Violin Plot

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.show()
```

Pengaturan pada Violin Plot

```python
plt.violinplot(
  data,
  showextrema=True,
  showmeans=True,
  showmedians=True,
  quantiles=[0.25, 0.5, 0.75] # untuk mengatur garis Q1, Q2, dan Q3
)

plt.title('Violin Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.show()
```

Horizontal Violin Plot

```python
plt.violinplot(
  data,
  showextrema=True,
  showmeans=True,
  showmedians=True,
  quantiles=[0.25, 0.5, 0.75], # untuk mengatur garis Q1, Q2, dan Q3
  vert=False # untuk mengatur horizontal
)

plt.title('Violin Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.show()
```

Multiple Violin Plot

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
```

```python
plt.violinplot(data)
plt.title('Violin Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.show()
```
