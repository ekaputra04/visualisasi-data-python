Pengaturan Figure Layouts dengan GridSpec

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
```

Simple 2x2 Layouts dengan Subplot

```python
siswa = ['Cecep', 'Joko', 'Wati', 'Ani']
uas = [68, 83, 96, 72]

fig, ax = plt.subplots(ncols=2, nrows=2, constrained_layout=True)

ax[0][0].bar(siswa, uas)
ax[0][1].pie(uas, labels=siswa)
ax[1][0].pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax[1][1].barh(siswa, uas)

plt.show()
```

Simple 2x2 Layout dengan GridSpec Cara 1

```python
fig = plt.figure(constrained_layout=True)

gs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

Simple 2x2 Layout dengan GridSpec Cara 2 (Lebih Efektif)

```python
fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

Rows Span dan Cols Span dengan GridSpec

```python
fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0, :]) # : artinya mengekspansi baris/ kolom
# ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

plt.show()
```

```python
fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0, :]) # : artinya mengekspansi baris/ kolom
# ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
# ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

```python
fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(ncols=3, nrows=3)

ax1 = fig.add_subplot(gs[0, :]) # : artinya mengekspansi baris/ kolom
ax1.set_title('gs[0, :]')
ax1.bar(siswa, uas)

ax2 = fig.add_subplot(gs[1, :-1])
ax2.set_title('gs[1, :-1]')
ax2.barh(siswa, uas)

ax3 = fig.add_subplot(gs[1:, -1])
ax3.set_title('gs[1:, -1]')
ax3.bar(siswa, uas)

ax4 = fig.add_subplot(gs[-1, 0])
ax4.set_title('gs[-1, 0]')
ax4.pie(uas, labels=siswa)

ax5 = fig.add_subplot(gs[-1, -2])
ax5.set_title('gs[-1, -2]')
ax5.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])

plt.show()
```
