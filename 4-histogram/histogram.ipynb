Histogram

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
mu = 100
sigma = 15

x = mu + sigma * np.random.randn(10000) # normal distribution
x
```

```python
x.shape
```

Histogram dengan pyplot Style

```python
plt.hist(
  x,
  bins=100,
  facecolor='g',
  alpha=0.75
)
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.title('Contoh Histogram')

plt.text(45, 300, '$\mu=100,\ \sigma=15$')
plt.grid()
plt.show()
```

Histogram dengan OO Style

```python
fig, ax = plt.subplots()
ax.hist(
  x,
  bins=100,
  facecolor='b',
  alpha=0.75
)

ax.set_xlabel('Sumbu x')
ax.set_ylabel('Sumbu y')
ax.set_title('Contoh Histogram')

ax.text(45, 300, '$\mu=100,\ \sigma=15$')
ax.grid()
plt.show()
```
